Script to generate the `ref_file.txt` with 4 variants and 1000 individuals, where the frequency of the variants follows a random binomial distribution with a MAF=0.1. Use `x.transpose` to change the matrix orientation

```sos
import numpy as np
output_haplos = open(ref_file, 'w')
haplos = np.random.binomial(1, 0.1, (1000, 4))
#haplos.transpose()
for row in haplos:
    for elem in row:
        print(elem, end='')
    print()
np.savetxt(output_haplos, haplos, fmt="%.0f", delimiter="")
output_haplos.close()
 
```

```sos
np.mean(haplos, axis=0) #uses the columns of the matrix
```
